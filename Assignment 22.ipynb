## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter Plot: Display the relationship between two numeric variables.

2. Bar Plot: Show the distribution of categorical data.

3. Box Plot: Visualize the distribution and detect outliers in numerical data.

4. Violin Plot: Combine features of box plots and kernel density plots, useful for showing the distribution of data.

5. Heatmap: Display the correlation matrix of numeric variables or visualize 2D data with color intensity.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title('Line Plot of "fmri" Dataset')
plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot two box plots
plt.figure(figsize=(10, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

plt.tight_layout()
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple="stack", bins=30, kde=True)
plt.title('Histogram of Price with Cut Variation')
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for 'species'
sns.pairplot(iris_data, hue='species')
plt.title('Pair Plot of Iris Dataset with Species Hue')
plt.show()
```
