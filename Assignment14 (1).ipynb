Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight, web framework written in Python. It is based on the Werkzeug toolkit and is designed to make web development as easy and flexible as possible.

Advantages :

Lightweight: Flask is a lightweight framework, which means that it has a small footprint and does not require a lot of resources to run. This makes it a good choice for small to medium-sized websites and web applications.

Flexible: Flask is a very flexible framework. It does not impose any specific conventions or patterns on developers, which gives them the freedom to develop their applications in the way that they want.

Extensible: Flask is a very extensible framework. There are many extensions available for Flask, which can add new features and functionality to Flask applications.

Easy to learn and use: Flask is relatively easy to learn and use, even for beginners. This is because Flask is a small and focused framework, and it has a well-written documentation.

Community support: Flask has a large and active community. This means that there are many resources available to help developers learn and use Flask, and there is a large pool of developers who can help with problems.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"


if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

The url_for() function is used in Flask for URL building. It takes a function name and optional keyword arguments as input, and returns the corresponding URL.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def index():
    return "This is the index page."

@app.route("/user/<username>")
def user(username):
    return f"Hello, {username}!"

if __name__ == "__main__":
    # Generate the URL for the index page.
    index_url = url_for("index")

    # Generate the URL for the user page with the username "john_doe".
    user_url = url_for("user", username="john_doe")

    print(index_url)
    print(user_url)
```
