Q1. Explain GET and POST methods.

GET and POST are two of the most common HTTP methods used to interact with web APIs. They are used to perform different operations on resources, such as retrieving data, creating new data, updating existing data, and deleting data.

GET is used to retrieve data from a resource. The data is returned in the response body of the HTTP request. GET requests are typically used to read data, such as retrieving the list of products in an online store or retrieving the details of a user profile.

POST is used to create new data or update existing data in a resource. The data to be created or updated is sent in the request body of the HTTP request. POST requests are typically used to create new accounts, submit orders, or update user profiles.

Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are used to generate dynamic web pages. They are written in HTML and can contain variables that will be replaced with values when the template is rendered.

The render_template() function is used to render a template and return the resulting HTML to the client. It takes two arguments: the name of the template file and a dictionary of variables that will be replaced in the template.
