```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Simple Scatter Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a scatter glyph
p.circle(x, y, size=10, color="navy", alpha=0.8)

# Show the plot in Jupyter notebook (optional, if using a Jupyter environment)
output_notebook()

# Display the plot
show(p)
```

Glyphs in Bokeh:

Glyphs are visual representations of data in Bokeh plots, such as lines, circles, rectangles, and more.
Adding Glyphs:

You can add glyphs to a Bokeh plot by calling the appropriate glyph function (e.g., circle(), line()) on the figure object.

```python
from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
p = figure(title="Example with Glyphs", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a circle glyph
p.circle(x, y, size=10, color="navy", alpha=0.8)

# Add a line glyph
p.line(x, y, line_width=2, line_color="red", line_dash="dashed")

# Show the plot
show(p)
```

## Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

1. Customize Axes:

Use x_axis_label and y_axis_label to set axis labels.
Adjust axis ticks, tick labels, and ranges using x_range, y_range, x_ticks, y_ticks, etc.

2. Customize Title:

Set the title of the plot using the title attribute of the figure.

3. Customize Legend:

Add legends to the plot by specifying the legend_label in the glyph functions.
Position and style the legend using legend_location, legend_label_text_font_size, etc.

## Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

Bokeh Server is a component of Bokeh that allows you to create interactive web applications with real-time updates.

Using Bokeh Server for Real-Time Updates:

1. Create a Bokeh App:

Define a Bokeh application by creating a function that modifies a Bokeh document.

2. Use curdoc() to Access Document:

Inside the Bokeh app function, use curdoc() to access the current Bokeh document.

3. Update Document for Real-Time Changes:

Update the document to reflect changes in real time. This can include updating plot data, adding or removing glyphs, etc.

4. Run Bokeh Server:

Run the Bokeh server with the bokeh serve command, specifying the script containing your Bokeh app.
