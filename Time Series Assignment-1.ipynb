Q4. How can time series forecasting be used in business decision-making, and what are some common challenges and limitations?

Q5. What is ARIMA modelling, and how can it be used to forecast time series data?

Q6. How do Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots help in identifying the order of ARIMA models?

Q7. What are the assumptions of ARIMA models, and how can they be tested for in practice?

Q8. Suppose you have monthly sales data for a retail store for the past three years. Which type of time series model would you recommend for forecasting future sales, and why?

Q9. What are some of the limitations of time series analysis? Provide an example of a scenario where the limitations of time series analysis may be particularly relevant.

Q10. Explain the difference between a stationary and non-stationary time series. How does the stationarity of a time series affect the choice of forecasting model?

## Answers

Q1. Time Series: A sequence of data points indexed in time order, often used to model trends, seasonality, and other patterns over time.

Applications: Sales forecasting, stock market analysis, weather prediction, resource management, anomaly detection (e.g., sensor data).

Q2. Patterns:

Trend: Long-term increase or decrease.

Seasonality: Cyclic variations at predictable intervals (e.g., monthly, yearly).

Cyclicity: Repeating patterns of ups and downs over longer periods.

Stationarity: Consistent statistical properties over time (mean, variance).

Identification: Visual inspection, statistical tests (e.g., seasonality tests).

Interpretation: Trends indicate long-term direction, seasonality helps in planning, cyclicity informs strategic decisions.

Q3. Preprocessing:

Handling missing values: Imputation (filling in missing data), deletion depending on extent and pattern.

Outlier removal: Correction, deletion, robust methods.

Scaling/normalization: Transforming data to a common scale if needed.

Q4. Business Forecasting:

Demand forecasting: Predicting future sales to optimize production, inventory, and marketing.

Financial forecasting: Projecting revenue, expenses, and cash flow for financial planning and budgeting.

Risk management: Identifying potential risks and opportunities by anticipating future trends.

Challenges: External factors can impact accuracy (e.g., economic changes, new regulations).

Limitations: Predictions are not guarantees, and model performance may degrade over time.

Q5. ARIMA: Autoregressive Integrated Moving Average model. A popular method for forecasting time series.

Autoregressive (AR): Uses past values of the series to predict future values.

Integrated (I): Differencing the data to achieve stationarity (if necessary).

Moving Average (MA): Incorporates past forecast errors to improve accuracy.

Q6. ACF & PACF:

ACF: Shows correlation of a series with its own lagged versions (identifies potential AR terms).

PACF: Shows correlation after removing influence of earlier lags (identifies pure AR terms).

Q7. ARIMA Assumptions:

Stationarity: Constant mean, variance, and autocorrelation over time.

Normality: Residuals (errors) are normally distributed.

No missing values or outliers.

Testing: Visual inspection of plots, statistical tests (e.g., Dickey-Fuller test for stationarity).

Q8. Monthly Sales: Seasonal ARIMA (SARIMA) model.

Considers monthly seasonality along with AR and MA components.

Suitable for capturing monthly sales patterns while accounting for trends and past values.

Q9. Limitations:

Past dependence: Relies on past data, may not capture unforeseen changes.

Stationarity requirement: Not always achievable for all datasets.

Example: Predicting future stock prices during a financial crisis. The model might not capture such drastic events not present in historical data.

Q10. Stationarity:

Stationary: Statistics (mean, variance) are constant over time.

Non-stationary: Statistics change over time (e.g., increasing trends).

Forecasting model:

Stationary: ARIMA, SARIMA, etc.

Non-stationary: Differencing or other transformation techniques may be needed before applying models like ARIMA.
